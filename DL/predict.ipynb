```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import TextVectorization
```

```python
model = tf.keras.models.load_model("./model/model_cleaned.keras")
```

```python
# Predict

# textInput = np.array(["Farine de blé 27%, chocolat au lait 18% (sucre, beurre de cacao, sel, gluten de blé, lait écrémé en poudre"])
# textInput = np.array([
#     "Corn flour, vegetable oil, salt, cheese powder (milk), flavors, color (annatto)."
#     ], dtype=object)

# numberInput = np.array([
#     [510.0,28.0,6.5,59.0,4.5,6.0,2.0,1.6,0.64]
# ])

textInput = np.array([
    "Tomatoes (60%), cucumbers (20%), onions, olive oil, lemon juice, sea salt, parsley."
    ], dtype=object)

numberInput = np.array([
    [45.0,1.2,0.2,8.5,4.0,1.1,1.8,0.3,0.12]
])

res = model.predict({
    # "text_input": np.array(["acid, sugar, milk, salt"], dtype=object),
    "text_input": textInput,
    # "numeric_input": np.array([[120, 5, 30, 0, 0, 0, 0, 0, 0]])     # example numbers
    "numeric_input": numberInput     # example numbers
})
print(res)
```

```python
class_names = ["Healthy", "Medium", "Not Healthy"]
```

```python
pred_index = np.argmax(res)

# Convert to label
pred_label = class_names[pred_index]

print("Prediction :", pred_label)
print("Probabilities :", res)
```
