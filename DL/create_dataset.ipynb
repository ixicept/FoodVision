## Import Library

```python
import pandas as pd
```

```python
columns_all = [
    "ingredients_text",
    "ingredients_tags",
    "ingredients_analysis_tags",
    "additives",
    "additives_tags",
    "additives_en",
    "additives_n",
    "allergens",
    "allergens_en",
    "nova_group",
    "nutriscore_score",
    "nutriscore_grade",
    "serving_size",
    "serving_quantity",
    "energy-kcal_100g",
    "fat_100g",
    "saturated-fat_100g",
    "carbohydrates_100g",
    "sugars_100g",
    "proteins_100g",
    "fiber_100g",
    "salt_100g",
    "sodium_100g"
]

columns = [
    "ingredients_text",
    "nutriscore_score",
    "nutriscore_grade",
    "energy-kcal_100g",
    "fat_100g",
    "saturated-fat_100g",
    "carbohydrates_100g",
    "sugars_100g",
    "proteins_100g",
    "fiber_100g",
    "salt_100g",
    "sodium_100g"
]

mapping = {
    "a": "Healthy",
    "b": "Healthy",
    "c": "Medium",
    "d": "Not Healthy",
    "e": "Not Healthy"
}

```

```python
DATASET_PATH = "./dataset/raw_data.csv"
```

## Load Dataset + Pre Processing

```python
# Read dataset from .csv
df = pd.read_csv(DATASET_PATH, sep="\t", nrows=100000)

# Filter columns to selected
df = df[columns]

# Drop empty data
df = df.dropna(how="all")
df = df[df["nutriscore_grade"].isin(["a", "b", "c", "d", "e"])]
df = df[df["ingredients_text"].str.len() > 0]

# Fill empty values with 0
df["energy-kcal_100g"] = df["energy-kcal_100g"].fillna(0)
df["fat_100g"] = df["fat_100g"].fillna(0)
df["saturated-fat_100g"] = df["saturated-fat_100g"].fillna(0)
df["carbohydrates_100g"] = df["carbohydrates_100g"].fillna(0)
df["sugars_100g"] = df["sugars_100g"].fillna(0)
df["proteins_100g"] = df["proteins_100g"].fillna(0)
df["fiber_100g"] = df["fiber_100g"].fillna(0)
df["salt_100g"] = df["salt_100g"].fillna(0)
df["sodium_100g"] = df["sodium_100g"].fillna(0)

# Classify healthy_level based on nutriscore_grade
df["healthy_level"] = df["nutriscore_grade"].map(mapping)

print(df.shape)
```

## Export Filtered Data

```python
df.to_csv("./dataset/filtered_data.csv", index=False)
```
